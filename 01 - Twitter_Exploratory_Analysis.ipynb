## Exploratory Data Analysis with MongoDB

In this exploratory data analysis Phase I of this Data Analysis project done by making use of the Aggregation Framework From MongoDB. We can substract and all the data related to the research questions proposed for this project. 

In a short description group specific values in a large dataset or document and which we can sort, match data to specific indexes. It's faster than the native Map-Reduce in MongoDB. Even if Map-reduce operations can be rewritten with aggregation pipeline for the purpose of this project i decided the design described in my report.


```python
import pymongo
from bson.son import SON
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["covidfall20"]
mycol = mydb["tweets"]

topUsersQuery = [
    { "$project": {
        "_id": 0,
        "entities.user_mentions" : 1
    }},
    { "$unwind": "$entities.user_mentions" },

    { "$group": {
        "_id": "$entities.user_mentions.screen_name",
        "count": { "$sum": 1 }
    }},

    {"$sort": SON([("count", -1)])}

]


data = pd.DataFrame(list(mydb.tweets.aggregate(topUsersQuery)))

data.head(25)

```

```python
tweetsCountByLang = [
{ "$group": {
"_id": '$lang',
"count": {"$sum": 1}
}},

{ "$match": {
"count": { "$gt": 100 }
}},

{"$sort": SON([("count", -1)])},

{ "$project": {
"language": '$_id',
"count": 1,
"_id": 0
}}
   
]

data = pd.DataFrame(list(mydb.tweets.aggregate(tweetsCountByLang)))

data.head(20)
```

```python

tweetsTopHashtags = [

    { "$project": {
       "_id": 0,
        "entities.hashtags" : 1
    }},
    { "$unwind": "$entities.hashtags" },

    { "$group" : {
        "_id": "$entities.hashtags.text",
        "count": { "$sum" : 1 }
    }},

  
   {"$sort": SON([("count", -1)])}

    ]
    
data = pd.DataFrame(list(mydb.tweets.aggregate(tweetsTopHashtags)))

data.head(20)



```
